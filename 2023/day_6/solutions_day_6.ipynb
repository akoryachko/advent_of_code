```python
def flist(nm):
    with open(f'{nm}.txt') as f:
        return [line.strip() for line in f.readlines()]
```

```python
def part1(fn):
    e1 = flist(fn)
    times = [int(n) for n in e1[0].split(':')[1].strip().split()]
    distances = [int(n) for n in e1[1].split(':')[1].strip().split()]

    num_of_ways = []
    for time, distance in zip(times, distances):
        num_of_ways.append([t*(time - t) for t in range(time) if t*(time - t) > distance])

    return [len(d) for d in num_of_ways]

def mul_list(x):
    res = 1
    for a in x:
        res = res*a
    return res

assert part1('example1') == [4, 8, 9]
assert mul_list(part1('example1')) == 288
```

```python
mul_list(part1('input'))
```
